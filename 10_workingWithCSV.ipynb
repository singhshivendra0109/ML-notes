# Introduction to Data Gathering

> **The Golden Rule:** Data is the fuel of Machine Learning. Even an excellent algorithm will perform poorly with insufficient data, while a "bad" algorithm can yield better results with a large amount of quality data.

## Why Data is Paramount

| Scenario | Algorithm Quality | Data Quality/Quantity | Result |
|----------|------------------|----------------------|--------|
| A | Excellent | Poor/Insufficient | ❌ Poor Performance |
| B | Average | Large & Quality | ✅ Better Results |

---

## Overview of Data Sources

There are **four key data acquisition methods** that will solve ~90% of common data acquisition problems:

| Method | Description | When to Use |
|--------|-------------|-------------|
| **CSV Files** | Most common and foundational format | Beginners, widespread availability |
| **JSON & SQL** | Universal data exchange format & database queries | APIs, structured databases |
| **Fetching from APIs** | Query data from website servers | Real-time data, external services |
| **Web Scraping** | Parse HTML to extract data | When no API/data source exists |

> 💡 **Note:** Other sources like Google Colab, data warehouses, and NoSQL databases exist, but mastering these four methods covers most use cases.

## Understanding CSV Files

> **CSV** stands for **Comma Separated Values** - a simple, universal format for storing tabular data.

### Structure of a CSV File

```
Name,Age,City,Salary
John,28,New York,50000
Alice,34,Boston,65000
Bob,22,Chicago,45000
```

| Component | Description |
|-----------|-------------|
| **Row** | Represents a single record |
| **Values** | Separated by commas within each row |
| **Header** | First row typically contains column names |

---

### CSV vs TSV

| Format | Separator | Example |
|--------|-----------|----------|
| CSV | Comma (`,`) | `John,28,NYC` |
| TSV | Tab (`\t`) | `John    28    NYC` |

# In-Depth: `pandas.read_csv()` Function

## Loading CSVs: Local vs Remote

### From Local Machine

```python
import pandas as pd

# Same directory
df = pd.read_csv('data.csv')

# Relative path
df = pd.read_csv('./folder/data.csv')

# Absolute path
df = pd.read_csv('C:/Users/data/file.csv')
```

---

### From Remote URL

```python
import requests
from io import StringIO

url = 'https://example.com/data.csv'
headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10.14; rv:66.0) Gecko/20100101 Firefox/56.0'}
req = requests.get(url, headers=headers)
data = StringIO(req.text)

pd.read_csv(data)
```

> ⚠️ **Note:** Some servers block requests without a proper `User-Agent` header. Always include headers to mimic a browser request.

## Key Parameters of `read_csv()`

---

### `sep` Parameter

> Specifies the delimiter that separates values. **Default is comma (`,`)**.

```python
# For tab-separated values (TSV)
df = pd.read_csv('file.tsv', sep='\t')

# For semicolon-separated
df = pd.read_csv('file.csv', sep=';')
```

| Issue | Cause | Solution |
|-------|-------|----------|
| Entire row in single column | Wrong separator | Specify correct `sep` |

---

### `names` Parameter

> Used when CSV file **lacks a header row** and Pandas incorrectly interprets the first data row as headers.

```python
# Provide custom column names
df = pd.read_csv('file.csv', names=['Name', 'Age', 'City', 'Salary'])
```

**Effect:** Pandas assigns provided names and treats the original first row as data.

---

### `index_col` Parameter

> Specify a column to use as the DataFrame's **index** instead of the default numerical index.

```python
# Use 'ID' column as index
df = pd.read_csv('file.csv', index_col='ID')

# Use first column as index
df = pd.read_csv('file.csv', index_col=0)
```

**Benefit:** Avoids redundant index columns in your data.

### `header` Parameter

> Addresses issues where the **header row is not the first row** or is incorrectly formatted.

```python
# Specify which row contains column names
df = pd.read_csv('file.csv', header=0)  # First row is header - default column names row
df = pd.read_csv('file.csv', header=1)  # Second row is header
df = pd.read_csv('file.csv', header=2)  # Third row is header
df = pd.read_csv('file.csv', header=None)  # No header present
```

---

### `usecols` Parameter

> Load only a subset of columns which are required from a large dataset.

```python
# Load specific columns only
df = pd.read_csv('file.csv', usecols=['Name', 'Salary'])

# Using column indices
df = pd.read_csv('file.csv', usecols=[0, 2, 4])
```

| Benefit | Description |
|---------|-------------|
| Memory Efficiency | Only selected columns are loaded |
| Faster Loading | Less data to process |
| Cleaner Data | Unneeded columns automatically dropped |

---

### `squeeze` Parameter

> When `True` and only **one column** is imported, converts DataFrame to a **Pandas Series**.

```python
# Result will be a Series instead of DataFrame
series = pd.read_csv('file.csv', usecols=['Age'], squeeze=True)
```

> 💡 **Note:** Less commonly used, but convenient for single-column data.
---

### `skiprows` Parameter

> Skip specific rows during import - useful for removing irrelevant or problematic rows.

```python
# Skip first 5 rows
df = pd.read_csv('file.csv', skiprows=5)

# Skip specific row indices (0-indexed)
df = pd.read_csv('file.csv', skiprows=[0, 2, 5])

# Skip using a custom function
df = pd.read_csv('file.csv', skiprows=lambda x: x % 2 == 1)  # Skip odd rows
```

---

### `nrows` Parameter

> **Essential for large datasets!** Limits the number of rows read from the file.

```python
# Load only first 1000 rows
df = pd.read_csv('large_file.csv', nrows=1000)
```

| Use Case | Benefit |
|----------|--------|
| Data Preview | Quick look at structure |
| Memory Management | Prevents RAM overload |
| Testing/Development | Work with sample data |

### `encoding` Parameter

> Addresses **UnicodeDecodeError** when file encoding differs from default UTF-8.

```python
# Common encodings
df = pd.read_csv('file.csv', encoding='utf-8')      # Default
df = pd.read_csv('file.csv', encoding='latin1')     # Latin-1
df = pd.read_csv('file.csv', encoding='ISO-8859-1') # Same as latin1
df = pd.read_csv('file.csv', encoding='cp1252')     # Windows encoding
```

| Error | Solution |
|-------|----------|
| `UnicodeDecodeError` | Try `encoding='latin1'` or `'ISO-8859-1'` |

---

### `on_bad_lines` Parameter

> Handles **ParserError** when rows have inconsistent number of columns.

```python
# Skip problematic rows (Pandas 1.3+)
df = pd.read_csv('file.csv', on_bad_lines='skip')

# Warn about bad lines
df = pd.read_csv('file.csv', on_bad_lines='warn')

# Old syntax (deprecated)
df = pd.read_csv('file.csv', error_bad_lines=False)
```

> ⚠️ **Note:** `on_bad_lines` replaced `error_bad_lines` in newer Pandas versions.

### `dtype` Parameter

> Explicitly set data types for columns, **overriding Pandas' inference**.

```python
# Specify column types
df = pd.read_csv('file.csv', dtype={
    'ID': int,
    'Name': str,
    'Price': float
})

# Optimize memory for integer columns
df = pd.read_csv('file.csv', dtype={'Age': 'int32', 'Score': 'float32'})
```

| Benefit | Description |
|---------|-------------|
| Memory Optimization | Use smaller data types |
| Type Safety | Ensure correct types from start |
| Avoid Conversion | No need for post-load `.astype()` |

---

### `parse_dates` Parameter

> Convert date columns from strings to **datetime objects** during import.

```python
# Parse single date column
df = pd.read_csv('file.csv', parse_dates=['date_column'])

# Parse multiple columns
df = pd.read_csv('file.csv', parse_dates=['start_date', 'end_date'])

# Combine multiple columns into single datetime
df = pd.read_csv('file.csv', parse_dates={'date': ['year', 'month', 'day']})
```

| Without `parse_dates` | With `parse_dates` |
|----------------------|--------------------|
| Dates as strings | Dates as `datetime64` |
| No date operations | Native date filtering & analysis |

### `converters` Parameter

> Apply a **custom function** to transform column values during loading.

```python
# Convert team names to short forms
team_short = {
    'Royal Challengers Bangalore': 'RCB',
    'Mumbai Indians': 'MI',
    'Chennai Super Kings': 'CSK'
}

df = pd.read_csv('ipl.csv', converters={
    'team': lambda x: team_short.get(x, x)
})

# Another example: clean phone numbers
df = pd.read_csv('contacts.csv', converters={
    'phone': lambda x: x.replace('-', '').replace(' ', '')
})
```

**Power:** Transform data at import time without post-processing!

---

### `na_values` Parameter

> Define custom string values to interpret as **missing values (NaN)**.

```python
# Treat specific strings as NaN
df = pd.read_csv('file.csv', na_values=['N/A', '-', 'missing', 'null'])

# Column-specific NA values
df = pd.read_csv('file.csv', na_values={
    'Age': ['unknown', '-1'],
    'Gender': ['not specified']
})
```

| Default NA Values | Custom NA Values |
|-------------------|------------------|
| `''`, `NA`, `NaN`, `null` | Whatever you specify |

### `chunksize` Parameter

> **Designed for extremely large datasets** that cannot fit entirely into RAM.

```python
# Returns an iterator, not a DataFrame
chunks = pd.read_csv('huge_file.csv', chunksize=5000)

# Process iteratively
for chunk in chunks:
    # Each chunk is a DataFrame with 5000 rows
    process(chunk)
```

### How `chunksize` Works

```
┌─────────────────────────────────────────┐
│           HUGE FILE (1M rows)           │
├─────────────────────────────────────────┤
│  Chunk 1  │ Rows 0 - 4999      (5000)   │
│  Chunk 2  │ Rows 5000 - 9999   (5000)   │
│  Chunk 3  │ Rows 10000 - 14999 (5000)   │
│    ...    │         ...                 │
│  Chunk N  │ Remaining rows              │
└─────────────────────────────────────────┘
```

### Example: Processing Large Files

```python
# Calculate mean of a column without loading entire file
total_sum = 0
total_count = 0

for chunk in pd.read_csv('sales.csv', chunksize=10000):
    total_sum += chunk['amount'].sum()
    total_count += len(chunk)

mean_amount = total_sum / total_count
```

| Benefit | Description |
|---------|-------------|
| Memory Safe | Never loads full file |
| Scalable | Handle files of any size |
| Iterative Processing | Apply transformations chunk by chunk |

## Summary: `read_csv()` Parameters at a Glance

| Parameter | Purpose | Example |
|-----------|---------|----------|
| `sep` | Change delimiter | `sep='\t'` |
| `names` | Provide column names | `names=['A', 'B']` |
| `index_col` | Set index column | `index_col='ID'` |
| `header` | Specify header row | `header=0` |
| `usecols` | Load specific columns | `usecols=['Name']` |
| `squeeze` | Convert to Series | `squeeze=True` |
| `skiprows` | Skip rows | `skiprows=[0, 2]` |
| `nrows` | Limit rows loaded | `nrows=1000` |
| `encoding` | Set file encoding | `encoding='latin1'` |
| `on_bad_lines` | Handle bad rows | `on_bad_lines='skip'` |
| `dtype` | Set column types | `dtype={'Age': int}` |
| `parse_dates` | Parse date columns | `parse_dates=['date']` |
| `converters` | Apply functions | `converters={'col': func}` |
| `na_values` | Define NA values | `na_values=['N/A']` |
| `chunksize` | Read in chunks | `chunksize=5000` |

## Key Takeaways

1. **Data > Algorithm** — Quality data is more important than a sophisticated algorithm
2. **Four methods solve 90%** of data acquisition: CSV, JSON/SQL, APIs, Web Scraping
3. **CSV is foundational** — Master `read_csv()` parameters for efficient data loading
4. **Memory matters** — Use `usecols`, `nrows`, and `chunksize` for large files
5. **Handle edge cases** — `encoding`, `on_bad_lines`, and `na_values` save debugging time
