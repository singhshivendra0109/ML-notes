# Machine Learning

Machine learning is a field of computer science that uses statistical techniques to give computer systems the ability to "learn" with data, without being explicitly programmed.

## Traditional Programming vs Machine Learning

| Aspect | Traditional Programming | Machine Learning |
|--------|------------------------|------------------|
| **Input** | Data + Program | Data + Output |
| **Process** | Computer executes rules | Computer learns patterns |
| **Output** | Output/Results | Program/Model |

# When is ML Better Than Traditional Programming?

### 1. When You Can't Write Cases for Everything
- **Example**: Email spam classifier
- Explicit rules would constantly need updating as new spam techniques emerge
- Impossible to anticipate every spam variation

### 2. When You Cannot Imagine the Number of Cases
- **Example**: Image classification
- Impossible to code for every variation of an object
- E.g., different dog breeds, angles, lighting conditions

### 3. Data Mining
- Extracting **hidden patterns** and information from data
- Reveals insights not apparent through simple analysis or graphs

## Symbolic AI / Expert Systems

- Early AI attempts involved building large knowledge systems with **"if-then" conditions**
- **Example**: Early chess programs

| Strengths | Limitations |
|-----------|-------------|
| Effective for **closed problems** with clear logic | Struggled with **open-ended problems** (image/voice recognition) |
| Rule-based, predictable behavior | Rules difficult to define for every variation |

> This limitation led to the rise of **Machine Learning**

---

## Machine Learning (ML)

- ML is a **subset of AI** that uses statistical techniques to find patterns in data
- **Key Difference from Symbolic AI**: No explicit programming or rule-based logic
### How ML Works:
Provide Data → System Learns Patterns → Makes Predictions

**Example**: Image Classification (Identifying a Dog)

| Traditional Programming | Machine Learning |
|------------------------|------------------|
| Write explicit rules for fur, ears, tail, etc. | Feed thousands of labeled dog images |
| Fails with new breeds or angles | Learns features automatically (edges, textures, shapes) |
| Constant manual updates needed | Improves with more data |

**How ML Does It:**
1. **Training**: Show the model thousands of labeled images (dogs vs. not dogs)
2. **Feature Learning**: Model automatically learns to recognize underlying features (fur patterns, ear shapes, snout structure)
3. **Prediction**: Given a new image, model predicts if it's a dog based on learned patterns

> **Modern Context**: Models like ResNet, EfficientNet, and Vision Transformers (ViT) can classify thousands of dog breeds with 90%+ accuracy, far surpassing hand-coded rules.
---
## Deep Learning (DL)
- DL is a **specialized subset of ML** using algorithms inspired by the human brain's neural networks
- Core unit: **Neuron/Perceptron** (biologically inspired, mathematically operated)
### Main Advantage Over Traditional ML:
**Automatic Feature Extraction** from raw data (no manual feature engineering needed)

| Traditional ML | Deep Learning |
|---------------|---------------|
| Requires **manual feature engineering** | **Automatically learns** features from raw data |
| Domain expertise needed to define features | Discovers complex patterns on its own |
| Features: edges, color histograms, SIFT, HOG | Features: hierarchical (edges → textures → shapes → objects) |

**How DL Extracts Features (Example: Image):**
Raw Pixels → [Layer 1: Edges] → [Layer 2: Textures] → [Layer 3: Parts] → [Layer 4: Objects]

> **Why This Matters**: This automatic feature extraction is why Deep Learning dominates in:
> - **Computer Vision** (CNNs, Vision Transformers)
> - **NLP** (Transformers, LLMs like GPT-4, LLaMA)
> - **Speech Recognition** (Whisper, Wav2Vec)
> 
> Manual feature engineering is now largely obsolete for complex tasks.

### DL Excels At:
- Image classification
- Speech recognition
- Text processing
- Complex problems where identifying relevant features is difficult

### Performance Characteristics:
- **High accuracy** on tasks where patterns are well-defined
- **Scalability** to handle large datasets
- **Flexibility** to adapt to different types of problems

### Performance vs Data: ML vs Deep Learning

| Aspect | Traditional ML | Deep Learning |
|--------|---------------|---------------|
| **Small Data** | Performs well, often better | May underperform (needs more data) |
| **Large Data** | Performance **plateaus** | Performance **keeps improving** |
| **Why?** | Limited model capacity | Larger models with more parameters can learn more complex patterns |

### Why Does This Happen?
**Traditional ML (Plateaus):**
- Fixed feature representations
- Model complexity is limited
- Cannot leverage additional data beyond a point

**Deep Learning (Scales):**
- More data → Better feature learning
- Larger models → More capacity to learn
- Neural networks can keep improving with more layers and data

### When to Use What?

| Scenario | Preferred Approach |
|----------|-------------------|
| Large datasets, complex features | **Deep Learning** |
| Smaller datasets | **Traditional ML** |
| Well-defined features already exist | **Traditional ML** |

> ⚠️ **Note**: DL is not always the best solution. For smaller datasets or problems where traditional ML performs well, ML might still be preferred.


# Types of Machine Learning

Machine Learning algorithms are categorized based on the **amount of supervision** they require during training.

## 1. Supervised Machine Learning

> **Definition:** Learning from data that has both **input** and **output** (labeled data). The goal is to find a relationship between inputs and outputs to predict future outcomes.

Labeled data means each input has a known correct output/answer attached to it.
### 1.1 Regression
- Used when the **output is numerical** (continuous values)
- **Example:** Predicting house prices based on:
  - Number of rooms
  - Number of bathrooms
  - Square footage

### 1.2 Classification
- Used when the **output is categorical** (discrete classes)
- **Examples:**
  - Email spam detection (Spam / Not Spam)
  - Weather prediction (Rain / No Rain)
  - Medical diagnosis (Disease / No Disease)

---

## 2. Unsupervised Machine Learning

> **Definition:** Learning from data that has **only inputs** (no labeled outputs). The goal is to discover hidden patterns or structures within the data.

### 2.1 Clustering
- **Groups similar data points** together into clusters without predefined categories
- **Examples:**
  - E-commerce: Grouping customers by behavior → personalized recommendations
  - Marketing: Identifying distinct market segments → targeted campaigns
  - Biology: Grouping genes with similar expression patterns
  - Location: Grouping GPS coordinates to identify popular areas/hotspots
### 2.2 Dimensionality Reduction
- **Reduces the number of input features** (columns) in a dataset while preserving important patterns
- **Benefits:**
  - Easier visualization of high-dimensional data
  - Faster algorithm performance
  - Reduced storage requirements
- **Examples:**
 - House Features: 
    - Before: rooms, bathrooms, kitchen size, balcony (4 features)
    - After: square feet area (1 feature that captures overall size)
  - MNIST: Reducing 784-dimensional handwritten digit images (28×28 pixels) to 2D/3D for visualization
  - Face Recognition: Thousands of pixels → ~100 key facial features
  - Genomics: Thousands of genes → key genetic patterns
  - Stock Market: Hundreds of stocks → few market indicators

### 2.3 Anomaly Detection
- **Identifies unusual or abnormal data points** that deviate significantly from the norm
- **Examples:**
  - Manufacturing defect detection
  - Credit card fraud detection
  - Network intrusion detection

### 2.4 Association Rule Learning
- **Discovers hidden relationships or associations** between items that frequently occur together (in large datasets)
- **Format:** "If X → Then Y" (People who buy X often buy Y)

- **Real-Life Examples:**
  - Supermarket: Beer & Diapers → Often bought together (famous retail discovery)
  - Fast Food: Burger + Fries → Suggest adding a drink
  - Netflix: Watched Movie A → Recommend Movie B
  - Amazon: Customers who bought this also bought...
  - Healthcare: Symptom A + Symptom B → Likely Disease X

- **Business Use:**
  - Product placement (put related items nearby)
  - Bundle deals & combo offers
  - Cross-selling recommendations

---

## 3. Semi-Supervised Learning

> **Definition:** A hybrid approach used when only a **small portion of data is labeled**, while the majority remains unlabeled.

### Why Semi-Supervised?
- **Problem:** Manual labeling is expensive (time + effort + expertise)
- **Opportunity:** Unlabeled data is cheap and abundant
- **Solution:** Learn from few labels, apply to many!

### How It Works:
| Step | Action |
|------|--------|
| 1️⃣ | Human labels **small sample** (e.g., 1%) |
| 2️⃣ | Model **learns patterns** from labeled data |
| 3️⃣ | Model **auto-labels** remaining 99% |
| 4️⃣ | Use expanded dataset for **improved learning** |

### Example: Google Photos  
*(Semi-Supervised + Clustering)*

| Step | What Happens | Details |
|------|--------------|---------|
| 1️⃣ | **Clustering** | System automatically groups similar faces into clusters (unsupervised - no human input) |
| 2️⃣ | **You Label** | You tag 1-2 photos: "This is Ayush" (human supervision on tiny sample) |
| 3️⃣ | **Propagation** | System applies "Ayush" label to ALL photos in that face cluster |
| 4️⃣ | **Result** | That particular cluster is tagged with minimal effort! |

> This combines **Unsupervised** (clustering faces) + **Supervised** (your labels) = **Semi-Supervised**
---

## 4. Reinforcement Learning

> **Definition:** An **agent** learns through **trial and error** by interacting with an environment. Unlike other types, there's no pre-existing dataset.

### Key Concepts:
| Term | Description |
|------|-------------|
| **Agent** | The learner/decision maker |
| **Environment** | What the agent interacts with |
| **Reward** | Positive feedback for desired actions |
| **Penalty** | Negative feedback for undesirable actions |
| **Policy** | The agent's "rulebook" or strategy |

### Learning Process:
1. Agent takes an action in the environment
2. Receives reward or penalty based on outcome
3. Gradually improves its policy over time
4. Eventually learns optimal behavior

###  Famous Example: AlphaGo (2016)
- AI developed by DeepMind
- Defeated the world champion in **Go** (an extremely complex board game)
- Learned by playing millions of games against itself

### ChatGPT (2022) & RLHF
- AI learns from **human feedback** as rewards/penalties
- Human rates responses: 👍 (reward) or 👎 (penalty)
- Model improves its responses based on this feedback

### Self-Driving Cars

| Agent | Environment | Reward | Penalty |
|-------|-------------|--------|---------|
| Car AI | Roads, traffic | Safe driving | Collisions, violations |

- **Actions:** Steer, accelerate, brake, lane change
- **Learning:** Tesla collects billions of miles → AI learns from real + simulated driving
- **Result:** Every Tesla contributes data → Fleet learns together!
