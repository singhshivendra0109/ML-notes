# Tensors: The Fundamental Data Structure of Machine Learning

> **Definition:** A tensor is a generalization of vectors and matrices - essentially a **multi-dimensional array** used to store numbers.

## Why Are Tensors Important?

- **Foundation of ML Libraries:** Leading libraries like **scikit-learn** and **TensorFlow** are built upon tensors as their basic data structure
- **Universal Data Representation:** All data in ML (images, text, audio, video) must be converted to tensors
- **The Name Says It All:** The very name "**TensorFlow**" underscores the significance of this concept - data literally *flows* through networks as tensors!

---

## Understanding Tensor Dimensions

### What is an Axis?

> An **axis** is a direction along which data is organized in a tensor. Each axis represents one dimension of the tensor.

| Tensor Type | Number of Axes | Axes Description |
|-------------|---------------|------------------|
| 0D (Scalar) | 0 | No axis - just a point |
| 1D (Vector) | 1 | Axis 0: along the elements |
| 2D (Matrix) | 2 | Axis 0: rows ↓, Axis 1: columns → |
| 3D Tensor | 3 | Axis 0: depth, Axis 1: rows, Axis 2: columns |

**Visual Example:**
```
1D: [1, 2, 3, 4]        → Axis 0 runs left to right

2D: [[1, 2, 3],         → Axis 0 runs top to bottom (rows)
     [4, 5, 6]]         → Axis 1 runs left to right (columns)

3D: [[[1, 2],           → Axis 0: depth (front to back) - which matrix
      [3, 4]],          → Axis 1: rows (top to bottom)
                        → Axis 2: columns (left to right)
     [[5, 6],
      [7, 8]]]          Shape: (2, 2, 2) - 2 matrices, each 2×2

    Visualized as stacked matrices:
    
         Matrix 0 (front)     Matrix 1 (back)
         ┌───┬───┐           ┌───┬───┐
         │ 1 │ 2 │           │ 5 │ 6 │
         ├───┼───┤           ├───┼───┤
         │ 3 │ 4 │           │ 7 │ 8 │
         └───┴───┘           └───┴───┘
            ↑                    ↑
         Axis 0 = 0          Axis 0 = 1
```

> 💡 **Remember:** The number of axes = Rank = Dimension of the tensor

### Tensor Types Overview

| Dimension | Name | Example | Description |
|-----------|------|---------|-------------|
| 0D | Scalar | `7` | A single number |
| 1D | Vector | `[1, 2, 3, 4]` | A list of numbers |
| 2D | Matrix | `[[1, 2], [3, 4]]` | A grid of numbers (rows × columns) |
| 3D | 3D Tensor | Cube of numbers | Collection of matrices |
| 4D | 4D Tensor | Images batch | Collection of 3D tensors |
| 5D | 5D Tensor | Videos batch | Collection of 4D tensors |

## 0D Tensors (Scalars)

> A scalar is a single number with **zero dimensions**.

**Examples:**
- `7`
- `20`
- `3.14`

**Properties:**
- Rank: 0
- Shape: `()`
- Size: 1

---

## 1D Tensors (Vectors)

> A vector is a **list of numbers** arranged along a single axis.

**Example:** `[1, 2, 3, 4]`

### ⚠️ Common Point of Confusion

| Term | Meaning |
|------|--------|
| **Tensor Dimension (1D)** | Number of axes = 1 |
| **Vector Dimension** | Number of elements (features) |

**Example:** `[8.9, 19, 0]`
- As a tensor: **1D tensor** (has 1 axis)
- As a vector: **3-dimensional vector** (has 3 elements/features)

### Practical Examples of 1D Tensors:

| Use Case | Example |
|----------|--------|
| Student Data (single row) | `[CGPA, IQ, State]` → `[8.5, 120, 1]` |
| Target Variable (single output) | `[Placed, Not Placed, Placed]` → `[1, 0, 1]` |
| Audio Signal | Amplitude values over time |

## 2D Tensors (Matrices)

> A matrix is a **collection of vectors** forming a grid-like structure with rows and columns.

**Structure:** Rows × Columns

### Visual Representation:
```
     Col1  Col2  Col3
Row1 [ 1    2     3  ]
Row2 [ 4    5     6  ]
Row3 [ 7    8     9  ]
```

### Common Use: Tabular Data

| Student | CGPA | IQ  | Placement |
|---------|------|-----|----------|
| 1       | 8.5  | 120 | 1        |
| 2       | 6.2  | 100 | 0        |
| 3       | 9.1  | 130 | 1        |

- Each **row** = 1D tensor (single student's data)
- Entire **table** = 2D tensor (all students' data)

**Shape:** (3, 4) → 3 students, 4 features

## Key Tensor Properties

Understanding these properties is essential for working with tensors:

| Property | Definition | Example (3×3 Matrix) |
|----------|------------|---------------------|
| **Rank** | Number of axes/dimensions | 2 |
| **Shape** | Number of items that is stored along each axis | (3, 3) |
| **Size** | Total number of elements | 9 (3 × 3) |

### Calculating Size
> **Size = Product of all shape elements**

For a Scaler: Size is always 1.

| Tensor Shape | Size Calculation | Total Elements |
|--------------|-----------------|---------------|
| (5) | 5 | 5 |
| (3, 4) | 3 × 4 | 12 |
| (2, 3, 4) | 2 × 3 × 4 | 24 |
| (10, 28, 28, 3) | 10 × 28 × 28 × 3 | 23,520 |

# Higher-Dimensional Tensors

---

## 3D Tensors

> A 3D tensor is a **collection of 2D matrices**, often visualized as a cube of numbers.

### Use Case 1: NLP / Text Data

**How text becomes a 3D tensor:**

| Level | Representation | Tensor Dimension |
|-------|---------------|------------------|
| Word | Numerical vector (vectorization) | 1D |
| Sentence | Collection of word vectors | 2D |
| Document/Batch | Collection of sentences | **3D** |

**Example:**
```
Sentence 1: "I love ML"     → [[0.2, 0.5], [0.8, 0.1], [0.3, 0.9]]
Sentence 2: "ML is fun"     → [[0.3, 0.9], [0.4, 0.2], [0.7, 0.5]]
Sentence 3: "Learn tensors" → [[0.1, 0.6], [0.9, 0.3]]

Batch of 3 sentences = 3D Tensor
Shape: (num_sentences, max_words, embedding_dim)
```

---

### Use Case 2: Time Series Data

**Example: Stock Price Data**

| Day | High | Low | Close |
|-----|------|-----|-------|
| 1   | 150  | 145 | 148   |
| 2   | 152  | 147 | 151   |
| ... | ...  | ... | ...   |

**Building up dimensions:**

| Data Scope | Tensor Dimension | Shape |
|------------|-----------------|-------|
| Single day's prices | 1D | (3,) |
| One year of prices | 2D | (365, 3) |
| Multiple years | **3D** | (years, 365, 3) |

**Shape:** `(num_years, num_days, num_features)`

## 4D Tensors

> 4D tensors are **prevalent in Computer Vision** for representing batches of images. These are vector of 3D tensors. 

### Understanding Image Representation

**Single Color Image = 3D Tensor:**
```
Height × Width × Color Channels
   ↓       ↓         ↓
  28   ×  28    ×    3 (RGB)
```

| Component | Description | Example |
|-----------|-------------|--------|
| Height | Rows of pixels | 224 |
| Width | Columns of pixels | 224 |
| Channels | Color depth (RGB=3, Grayscale=1) | 3 |

---

### Batch of Images = 4D Tensor

**Structure:**
```
(Number of Images × Height × Width × Channels)
         ↓            ↓       ↓        ↓
        32      ×   224   ×  224   ×   3
```

| Shape Component | Meaning |
|-----------------|--------|
| 1st dimension | Batch size (how many images) |
| 2nd dimension | Image height (pixels) |
| 3rd dimension | Image width (pixels) |
| 4th dimension | Color channels |

**Example:** Training batch of 32 RGB images of size 224×224
- Shape: `(32, 224, 224, 3)`
- Size: 32 × 224 × 224 × 3 = **4,816,896 values!**

## 5D Tensors

> 5D tensors are used for **video data** - the highest dimension commonly encountered. These are matrix of 4D tensors.

### Understanding Video Representation

**Key Insight:** A video is a sequence of frames (images)!

| Data Type | Tensor Dimension | Shape |
|-----------|-----------------|-------|
| Single Image | 3D | (H, W, C) |
| Single Video | 4D | (Frames, H, W, C) |
| **Batch of Videos** | **5D** | **(Videos, Frames, H, W, C)** |

### 5D Tensor Structure:

```
(Num Videos × Num Frames × Height × Width × Channels)
     ↓           ↓          ↓        ↓        ↓
     8     ×    30     ×   480   ×  640   ×   3
```

### Storage Requirements: The BIG Challenge!

**Example Calculation:**

| Parameter | Value |
|-----------|-------|
| Videos | 1 |
| Frames (30fps × 60sec) | 1,800 |
| Height | 1080 |
| Width | 1920 |
| Channels (RGB) | 3 |

**Total Values:** 1 × 1,800 × 1080 × 1920 × 3 = **11,197,440,000** (~11 billion!)

**Storage (8 bits per value):** ~11 GB for just 1 minute of video!

> ⚠️ **This is why video compression formats (like .mp4, .h264) are crucial!** They reduce storage by 100x or more through clever encoding.

### Real-World Applications of 5D Tensors:

| Application | Description |
|-------------|------------|
| Video Classification | Identifying actions in videos |
| Video Prediction | Predicting future frames |
| Self-Driving Cars | Processing video streams |
| Surveillance | Anomaly detection in CCTV footage |

## Summary: Tensor Dimensions at a Glance

| Dimension | Name | Common Use Case | Example Shape |
|-----------|------|-----------------|---------------|
| 0D | Scalar | Single value | `()` |
| 1D | Vector | Features, sequence | `(128,)` |
| 2D | Matrix | Tabular data | `(1000, 10)` |
| 3D | 3D Tensor | Text, Time series | `(32, 100, 300)` |
| 4D | 4D Tensor | Image batches | `(32, 224, 224, 3)` |
| 5D | 5D Tensor | Video batches | `(8, 30, 480, 640, 3)` |

---

## Key Takeaways

1. **Tensors are the universal language of ML** — all data must be converted to tensor form
2. **Dimension = Number of axes**, not the size along any axis
3. **Shape describes the structure**: how many elements along each axis
4. **Higher dimensions = More complex data types** (images → videos → batches)
5. **Storage grows exponentially** with dimensions — compression is essential for video!

> 💡 **Pro Tip:** When working with any ML framework, always check `.shape` to understand your tensor's structure. Shape mismatches are one of the most common errors in deep learning!
