# Framing a Business Problem as a Machine Learning Problem

Framing a machine learning problem is a **critical leadership skill** for data scientists. It ensures efficient project execution and avoids costly reworks - a key factor in career progression.

---

## 1. Defining the Type of Problem

> **Goal:** Identify the correct problem type (classification, regression, clustering, etc.) to guide your modeling approach.

### Example: Netflix Customer Churn Prediction

| Initial Approach | Refined Approach |
|-----------------|------------------|
| **Classification:** Will customer leave? (Yes/No) | **Regression:** What is the *probability* of leaving? |

### Why Probability (Regression) is Better:
- Enables **differentiated actions** based on churn likelihood
- High probability → Offer larger discounts
- Medium probability → Send personalized recommendations
- Low probability → Standard engagement

> ⚠️ **Key Insight:** A nuanced problem framing (regression over binary classification) allows for more strategic business decisions.

## 2. Investigating Current Solutions

> **Goal:** Understand what already exists before building something new.

### Why This Matters:
- Avoid reinventing the wheel
- Draw inspiration from previous approaches
- Understand which factors were previously considered
- Identify gaps or areas for improvement

### Questions to Ask:
| Question | Purpose |
|----------|--------|
| Are there existing models addressing this problem? | Baseline comparison |
| What features were used previously? | Feature engineering ideas |
| What were the limitations of past solutions? | Improvement opportunities |
| How was success measured before? | Metric alignment |

## 3. Gathering Data

> **Goal:** Clearly define what data is needed, what columns it should contain, and how to extract it.

### Example: Netflix Churn Prediction Data Requirements

| Data Point | Description | Why It's Important |
|------------|-------------|-------------------|
| **Watch Time** | Hours spent watching content | High engagement indicator |
| **Browsing Time** | Time spent browsing without watching | May indicate decision fatigue / Not getting their liked content|
| **Search Queries** | What users are searching for | Shows content expectations |
| **Search Results** | What the system returned | Indicates content availability |
| **Abandoned Shows** | Shows started but not finished | Dissatisfaction signal , Content is not tempting / interesting |

### Collaboration is Key:
- Work closely with **Data Engineers** to extract data from databases
- Ensure data quality and completeness
- Define data freshness requirements
- Document data sources and transformations

> 💡 **Tip:** Create a clear data specification document before extraction to avoid back-and-forth communication delays.

## 4. Defining Metrics to Measure Success

> **Goal:** Establish clear, measurable criteria to evaluate model performance.

### For Churn Prediction:

| Metric | Description |
|--------|-------------|
| **Predicted vs Actual Churn Rate** | Does prediction match reality? |
| **Precision** | Of predicted churners, how many actually churned? |
| **Recall** | Of actual churners, how many did we identify? |
| **Business Impact** | Did targeted interventions reduce churn? |

### Key Questions:
1. Were the customers we targeted the ones who actually churned?
2. Did our interventions (discounts, recommendations) work?
3. What's the ROI of the model compared to no intervention?

> ✅ **Best Practice:** Define both **model metrics** (accuracy, precision, recall) AND **business metrics** (revenue saved, churn reduction %).

## 5. Online vs Batch Learning

> **Goal:** Decide how the model should learn and update over time.

### Comparison:

| Aspect | Batch Learning | Online Learning |
|--------|---------------|----------------|
| **Training** | Retrained periodically (daily/weekly) | Learns continuously from new data |
| **Data** | Uses historical data | Uses real-time/streaming data |
| **Adaptability** | Slower to adapt to changes | Quickly adapts to new patterns |
| **Complexity** | Simpler to implement | More complex infrastructure needed |
| **Use Case** | Stable patterns | Rapidly changing environments |

### For Churn Rate (Volatile Concept):

| Consideration | Decision |
|--------------|----------|
| Ideal approach | Online Learning (churn patterns evolve) |
| Practical approach | Periodic Batch Retraining |

> 📌 **Reality Check:** While online learning seems ideal for volatile concepts like churn, practical limitations (infrastructure, cost, complexity) often necessitate periodic batch retraining.

## 6. Checking Assumptions

> **Goal:** Verify critical assumptions before full-scale implementation to avoid costly failures.

### Key Assumptions to Verify:

| Assumption | Verification Method |
|------------|--------------------|
| **Feature Availability** | Check if chosen features are actually accessible in production |
| **Data Quality** | Validate data completeness, accuracy, and freshness |
| **Geographic Applicability** | Test if model works across different regions/demographics |
| **Temporal Stability** | Ensure patterns hold over time (no concept drift) |
| **Scalability** | Verify the solution can handle production traffic |

### Common Pitfalls:
- Features available in training but not in production
- Model trained on one demographic, deployed to another
- Assumptions about data freshness that don't hold
- Infrastructure limitations not considered during development

> ⚠️ **Warning:** Failing to verify assumptions early leads to costly reworks and project delays.

## Quick Summary: Steps to Frame an ML Problem

| Step | Action | Key Question |
|------|--------|-------------|
| 1️⃣ | **Define Problem Type** | Classification, Regression, or something else? |
| 2️⃣ | **Investigate Existing Solutions** | What's already been tried? |
| 3️⃣ | **Gather Data** | What data do we need and how to get it? |
| 4️⃣ | **Define Success Metrics** | How do we know if it works? |
| 5️⃣ | **Choose Learning Approach** | Online or Batch? |
| 6️⃣ | **Verify Assumptions** | Are our assumptions valid? |

---

> 🎯 **Key Takeaway:** Problem framing is not just a technical exercise - it's a **leadership skill** that separates good data scientists from great ones. Getting this right ensures efficient execution and avoids costly reworks.
