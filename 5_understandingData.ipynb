# Understanding Your Data: Essential Questions for Data Quality Analysis

> **Key Insight:** Quality data is fundamental for gaining actionable insights. This process involves exploring data characteristics, patterns, and relationships.

Before building any ML model, you must thoroughly understand your dataset. This notes covers the **7 fundamental questions** to ask when first encountering a dataset.

---

## Essential Data Exploration Commands

| Question | Command | What It Reveals |
|----------|---------|------------------|
| How big is the data? | `df.shape` | (rows, columns) |
| How does the data look? | `df.sample(5)` | 5 random rows |
| What are the data types? | `df.info()` | Column types, memory usage |
| Any missing values? | `df.isnull().sum()` | Count per column |
| Mathematical summary? | `df.describe()` | Statistics for numerical columns |
| Any duplicates? | `df.duplicated().sum()` | Total duplicate rows |
| Column correlations? | `df.corr()` | Pearson correlation matrix |

## Question 1: How Big is the Data?

> Understanding the **scale** of your dataset is the first step - knowing the number of rows (samples) and columns (features).

### Command:
```python
df.shape
```

### What It Returns:

| Output | Meaning |
|--------|--------|
| `(1000, 15)` | 1000 rows (samples), 15 columns (features) |
| First value | Number of **observations/records** |
| Second value | Number of **features/variables** |

### Why It Matters:

- **Resource Planning:** Large datasets require more memory and processing power
- **Model Selection:** Some algorithms don't scale well with large datasets
- **Sampling Decision:** Decide if you need to work with a sample for initial exploration

> 💡 **Pro Tip:** A dataset with too few rows relative to columns can lead to overfitting!

## Question 2: How Does the Data Look?

> Get a **visual glimpse** of your data to understand its structure and content.

### Two Approaches:

| Method | Command | Use Case |
|--------|---------|----------|
| Top Rows | `df.head()` | Quick peek at first 5 rows |
| Bottom Rows | `df.tail()` | Check last 5 rows |
| **Random Sample** | `df.sample(5)` | **Recommended** - unbiased view |

### Why `df.sample(5)` is Better:

```python
# Avoid this potential bias:
df.head()  # First rows might be sorted or grouped unnaturally

# Prefer this for unbiased exploration:
df.sample(5)  # Random selection represents the whole dataset
```

### ⚠️ Common Pitfall

**Problem:** If data is sorted (e.g., by date, ID, or category), `head()` shows only one type of data.

**Example:**
```
If Titanic data is sorted by class:
✗ df.head() → Shows only 1st class passengers
✓ df.sample(5) → Shows mix of all classes
```

> 💡 **Best Practice:** Always use `sample()` for initial data exploration to avoid biased first impressions!

## Question 3: What is the Data Type of Columns?

> Understanding **data types** is crucial for operations, memory optimization, and correct analysis.

### Command:
```python
df.info()
```

### What It Reveals:

| Information | Description |
|-------------|-------------|
| Column names | List of all features |
| Non-null counts | Number of valid values per column |
| **Data types** | int64, float64, object, etc. |
| **Memory usage** | Total RAM consumption |

Null counts=Total counts-Non Null counts

### Common Data Types in Pandas:

| Type | Description | Example |
|------|-------------|---------|
| `int64` | Integer numbers | Age: 25, 30, 45 |
| `float64` | Decimal numbers | Salary: 50000.50 |
| `object` | Text/strings | Name: "John", "Alice" |
| `bool` | True/False | IsActive: True |
| `datetime64` | Dates and times | 2024-01-15 |

---

### Memory Optimization Opportunity

**Common Issue:** Numerical columns stored as `float64` when they should be `int`.

**Example:**
```python
# Age stored as float (wastes memory):
Age: [25.0, 30.0, 45.0]  → float64 uses 8 bytes per value

# Convert to integer (saves memory):
df['Age'] = df['Age'].astype('int32')  → int32 uses 4 bytes per value
```

### Memory Savings Table:

| Data Type | Bytes per Value | Use Case |
|-----------|-----------------|----------|
| `int8` | 1 | Small integers (-128 to 127) |
| `int16` | 2 | Medium integers (-32K to 32K) |
| `int32` | 4 | Standard integers |
| `int64` | 8 | Large integers |
| `float32` | 4 | Decimals (less precision) |
| `float64` | 8 | Decimals (high precision) |

> ⚠️ **Important:** In large datasets (millions of rows), optimizing data types can reduce memory usage by **50-75%**!

## Question 4: Are There Any Missing Values?

> Missing values can **significantly impact** your analysis and model performance. Identifying them is critical!

### Commands:

| Method | Command | Output |
|--------|---------|--------|
| Quick check | `df.info()` | Shows non-null counts |
| **Exact count** | `df.isnull().sum()` | Missing values per column |
| Percentage | `df.isnull().mean() * 100` | % missing per column |
| Total missing | `df.isnull().sum().sum()` | Total across dataset |

### Understanding the Output:

```python
# Example output of df.isnull().sum():
PassengerId      0      # No missing values ✓
Age             177     # 177 missing values !
Cabin           687     # 687 missing values ⚠️
Embarked          2     # 2 missing values
```

### Missing Value Categories:

| Missing % | Severity | Typical Action |
|-----------|----------|----------------|
| 0-5% | Low | Impute (fill with mean/median/mode) |
| 5-20% | Moderate | Impute carefully or drop if not critical |
| 20-50% | High | Consider dropping column or advanced imputation |
| >50% | Critical | Likely drop the column |

---

### Common Handling Strategies:

| Strategy | Code | When to Use |
|----------|------|-------------|
| Drop rows | `df.dropna()` | Few missing values |
| Drop column | `df.drop('col', axis=1)` | >50% missing |
| Fill with mean | `df['col'].fillna(df['col'].mean())` | Numerical, normally distributed |
| Fill with median | `df['col'].fillna(df['col'].median())` | Numerical, skewed data |
| Fill with mode | `df['col'].fillna(df['col'].mode()[0])` | Categorical data |

If you have a DataFrame with a column called 'Age', you would replace 'col' with 'Age'

> 💡 **Pro Tip:** Always investigate **why** data is missing before deciding how to handle it!

## Question 5: How Does the Data Look Mathematically?

> Get a **statistical summary** of numerical columns to understand distributions and spot anomalies.

### Command:
```python
df.describe()
```

### What It Returns:

| Statistic | Symbol | Description |
|-----------|--------|-------------|
| **count** | n | Number of non-null values |
| **mean** | μ | Average value |
| **std** | σ | Standard deviation (spread) |
| **min** | - | Minimum value |
| **25%** | Q1 | First quartile (25th percentile) |
| **50%** | Q2 | Median (middle value) |
| **75%** | Q3 | Third quartile (75th percentile) |
| **max** | - | Maximum value |

A percentile tells you what percentage of data falls below a certain value.

### Example Output:

```
           Age      Fare     SibSp     Parch
count   714.00    891.00    891.00    891.00
mean     29.70     32.20      0.52      0.38
std      14.53     49.69      1.10      0.81
min       0.42      0.00      0.00      0.00
25%      20.12      7.91      0.00      0.00
50%      28.00     14.45      0.00      0.00
75%      38.00     31.00      1.00      0.00
max      80.00    512.33      8.00      6.00
```

---

### 🔍 What to Look For:

| Pattern | Indication | Example |
|---------|------------|---------|
| **Large std vs mean** | High variability | Fare: mean=32, std=49 |
| **Min/Max far from mean** | Potential outliers | Fare: max=512, mean=32 |
| **Count < total rows** | Missing values | Age: 714 vs 891 total |
| **Mean ≠ Median (50%)** | Skewed distribution | If mean > median: right-skewed |
| **Same 25%, 50%, 75%** | Low variance | SibSp: all quartiles near 0 |

> ⚠️ **Anomaly Detection:** Extreme values in min/max often indicate data entry errors or genuine outliers that need investigation!

### For Categorical Columns:

```python
df.describe(include='object')  # Include only object (text) columns
df.describe(include='all')     # Include all column types
```

💡 Use Case: Use include='object' to quickly check categorical distributions - like how many unique categories exist and which is most common!

## Question 6: Are There Any Duplicate Values?

> Duplicate rows can **skew your analysis** and machine learning algorithms. Always identify and consider removing them!

### Commands:

| Task | Command | Output |
|------|---------|--------|
| Count duplicates | `df.duplicated().sum()` | Number of duplicate rows |
| View duplicates | `df[df.duplicated()]` | Show duplicate rows |
| Check specific columns | `df.duplicated(subset=['col1', 'col2'])` | Duplicates based on subset |
| Remove duplicates | `df.drop_duplicates()` | DataFrame without duplicates |

### Understanding `duplicated()`:

```python
# How it works:
df.duplicated()  # Returns True for duplicate rows (keeps first occurrence as False)

# Keep options:
df.duplicated(keep='first')  # Keep first, mark rest as duplicates (default)
df.duplicated(keep='last')   # Keep last, mark rest as duplicates
df.duplicated(keep=False)    # Mark ALL duplicates as True

keep=False is for data investigation, not deletion. It helps you understand the full picture before making decisions!
```

### Example Scenario:

```
Original Data:
   ID   Name    Age
0   1   John     25   ← First occurrence
1   2   Alice    30
2   1   John     25   ← Duplicate! (exact match)
3   3   Bob      35

df.duplicated().sum() → 1 (one duplicate found)
```

---

### Why Duplicates Are Problematic:

| Problem | Impact |
|---------|--------|
| **Biased statistics** | Mean, median skewed towards duplicated values |
| **Model overfitting** | Algorithm learns duplicated patterns too strongly |
| **Inflated metrics** | Accuracy appears better than reality |
| **Train-test leakage** | Same record in both sets = invalid evaluation |

> 💡 **Best Practice:** Always check for and remove duplicates before splitting data into train/test sets!

## Question 7: How is the Correlation Between Columns?

> **Correlation** helps understand relationships between numerical columns, especially how input features relate to the target variable.

### Command:
```python
df.corr()  # Calculates Pearson correlation coefficient
```

### Understanding Correlation Values:

| Correlation | Range | Meaning | Visual |
|-------------|-------|---------|--------|
| **Strong Positive** | +0.7 to +1.0 | Variables move together ↑↑ | ／ |
| **Moderate Positive** | +0.3 to +0.7 | Somewhat related ↑↗ | ⟋ |
| **Weak/None** | -0.3 to +0.3 | Little to no relationship | — |
| **Moderate Negative** | -0.7 to -0.3 | Inverse relationship ↑↘ | ⟍ |
| **Strong Negative** | -1.0 to -0.7 | Variables move oppositely ↑↓ | \ |

### Visual Representation:

```
 -1.0         0         +1.0
  │←─────────┼─────────→│
  │          │          │
Strong    No/Weak    Strong
Negative  Correlation Positive
  ↓↑         —          ↑↑
```

---

### Example Correlation Matrix:

```
              Survived    Pclass      Age     Fare    PassengerId
Survived         1.00     -0.34    -0.08     0.26          -0.01
Pclass          -0.34      1.00    -0.37    -0.55           0.03
Age             -0.08     -0.37     1.00     0.10           0.04
Fare             0.26     -0.55     0.10     1.00           0.01
PassengerId     -0.01      0.03     0.04     0.01           1.00
```

### 🔍 Key Insights from Correlation:

| Finding | Insight |
|---------|---------|
| `Survived` ↔ `Pclass`: -0.34 | Lower class → lower survival (negative) |
| `Survived` ↔ `Fare`: +0.26 | Higher fare → higher survival (positive) |
| `Survived` ↔ `PassengerId`: -0.01 | **No correlation** (near zero) |
| `Pclass` ↔ `Fare`: -0.55 | Higher class (1st) paid more |

---

### Correlation of a Specific Column with All Others:

| Task | Command |
|------|--------|
| Correlation of one column with all | `df.corr()['ColumnName']` |
| Sorted by strength | `df.corr()['ColumnName'].abs().sort_values(ascending=False)` |
| Top 5 most correlated | `df.corr()['ColumnName'].abs().nlargest(5)` |

This is useful for **feature selection** - quickly identify which features are most related to your target variable!

### Applications in ML:

| Use Case | Action |
|----------|--------|
| **Feature Selection** | Keep features with high correlation to target |
| **Remove Irrelevant** | Drop features with ~0 correlation (e.g., PassengerId) |
| **Multicollinearity** | If two features are highly correlated (>0.8), consider dropping one |
| **Feature Engineering** | Combine correlated features or create ratios |

### Visualization:

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create a heatmap for better visualization
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', center=0)
plt.title('Correlation Heatmap')
plt.show()
```

> ⚠️ **Important:** Correlation only measures **linear** relationships. Two variables can be strongly related non-linearly but show low correlation!
