# Exploratory Data Analysis: Univariate Analysis

> **Definition:** Univariate analysis is the simplest form of data analysis where we analyze **one variable (column) at a time** to understand its characteristics, patterns, and distributions.

## What is EDA?

**Exploratory Data Analysis (EDA)** is the preliminary analysis of data to:
- Discover **relationships**, **trends**, and **patterns**
- Understand data quality and structure
- Prepare for more advanced analysis or modeling

---

## Types of EDA Analysis

### By Method:
| Method | Description | Examples |
|--------|-------------|----------|
| **Graphical** | Visual representations | Histograms, Bar charts, Box plots |
| **Non-Graphical** | Statistical summaries | Mean, Median, Standard Deviation |

### By Number of Variables:
| Type | Variables | Focus |
|------|-----------|-------|
| **Univariate** | 1 | Single variable analysis |
| **Bivariate** | 2 | Relationship between two variables |
| **Multivariate** | 3+ | Complex relationships |

---

## Univariate Analysis: Deep Dive

> "**Uni**" means one, and "**variate**" means variable. The goal is to **derive, define, summarize, and analyze patterns** within each variable separately.

### Why Perform Univariate Analysis?

1. **Understand Distribution** - How is the data spread?
2. **Detect Outliers** - Are there unusual values?
3. **Find Central Tendency** - What's the typical value?
4. **Measure Spread** - How much do values vary?
5. **Identify Patterns** - Any skewness or multimodality?

# Understanding Data Types

> ⚠️ **Critical:** Before performing univariate analysis, you MUST understand the data types. The approach differs significantly between numerical and categorical data!

---

## Two Main Data Types

| Data Type | Description | Examples | Analysis Approach |
|-----------|-------------|----------|-------------------|
| **Numerical (Quantitative)** | Represents measurable quantities | Height, Age, Price, Temperature | Histograms, Box plots, Mean, Std |
| **Categorical (Qualitative)** | Represents categories or groups | Gender, Country, Color, Grade | Count plots, Pie charts, Frequency tables |

---

## Numerical Data Subtypes

| Subtype | Description | Example |
|---------|-------------|--------|
| **Continuous** | Can take any value in a range | Height: 170.5 cm, Temperature: 36.7°C |
| **Discrete** | Only specific values (usually integers) | Number of students: 25, Age: 19 |

## Categorical Data Subtypes

| Subtype | Description | Example |
|---------|-------------|--------|
| **Nominal** | No inherent order | Gender, Country, Platform |
| **Ordinal** | Has a natural order | Education Level (High School < Undergraduate < Graduate) |

---
Categorical data **can have numerical values as input**. The distinction between numerical and categorical data isn't about whether the input *looks* like a number-it's about **how those numbers are interpreted**.
## Examples

| Example | Values | Why it's Categorical |
|---------|--------|---------------------|
| **Rating** | 1, 2, 3, 4, 5 | Fixed set of discrete options, not continuous |
| **Education Level** | 1, 2, 3, 4 (High School → PhD) | Numbers are just labels/codes |
| **Product Size** | 1, 2, 3 (Small, Medium, Large) | Finite, fixed categories |
| **Zip/PIN Codes** | 110001, 400001 | Numbers representing locations (no math meaning) |
| **Year of Birth** | 1990, 1995, 2000 | Can be treated as categories for grouping |

---

## The Key Distinction

1. **Meaning**: Can you perform meaningful arithmetic?  
   - "Rating 4 + Rating 2 = Rating 6" makes **no sense** → Categorical

2. **Fixed Values**: Is there a limited, predefined set of possible values?

3. **Order**:
   - **Ordinal**: Has natural order (e.g., ratings 1-5, education levels)
   - **Nominal**: No inherent order (e.g., zip codes, colors)

---

### 💡 Key Insight

```
Numerical Data → Focus on DISTRIBUTION & CENTRAL TENDENCY
                 Tools: Histograms, Box plots, Mean, Median, Std

Categorical Data → Focus on FREQUENCY & PROPORTIONS
                   Tools: Count plots, Pie charts, Value counts
```

# Univariate Analysis for Categorical Data

> For categorical data, the primary goal is to understand the **frequency distribution** of each category.

---

## Key Questions to Answer:

1. **What categories exist?** — Get unique values
2. **How many of each?** — Count frequency
3. **What's the proportion?** — Calculate percentages
4. **Is there imbalance?** — Compare category sizes

---

## Tools for Categorical Analysis

| Tool | Purpose | When to Use |
|------|---------|-------------|
| `value_counts()` | Get frequency of each category | Quick summary |
| **Count Plot** | Visualize frequency as bars | Compare category counts |
| **Pie Chart** | Visualize proportions as slices | Show percentages |


# value_counts()

> `value_counts()` is a pandas method that returns the frequency of each unique value in a column, sorted by count (descending).

# Basic Syntax

```python
df['column_name'].value_counts()
```

---
# Plotting value_counts() Directly

> Chain `.plot()` to quickly visualize frequency distribution.

## Syntax:

```python
df['column_name'].value_counts().plot(kind='bar')
```

## Plot Options:

| kind | Description |
|------|-------------|
| `'bar'` | Vertical bar chart |
| `'barh'` | Horizontal bar chart |
| `'pie'` | Pie chart |

## Pie Chart with value_counts()

> Use `kind='pie'` to create a pie chart showing proportions.

### Syntax:

```python
df['column_name'].value_counts().plot(kind='pie', autopct='%.2f', colors=['#ff6b6b', '#4ecdc4'], explode=[0.02, 0.02], shadow=True, startangle=90)
```
### Key Parameters:

| Parameter | Description | Example |
|-----------|-------------|---------|
| `autopct` | Format for percentage labels | `'%.2f'`, `'%1.1f%%'` |
| `colors` | Custom colors for slices | `['#ff6b6b', '#4ecdc4']` |
| `explode` | Separate slices from center | `[0.02, 0.02]` |
| `shadow` | Add shadow effect | `True` |
| `startangle` | Rotation start angle | `90` |


## 1. Count Plot (Bar Chart)

> A **count plot** displays the frequency of each category as vertical (or horizontal) bars.

### Advantages:
- Easy to compare absolute counts
- Works well with many categories
- Clear visual hierarchy

### Syntax:
```python
sns.countplot(data=df, x='column_name')
```

````python
# Count Plot for Survived
plt.figure(figsize=(8, 5))
ax = sns.countplot(data=df, x='survived', palette='coolwarm')

# Add count labels on bars
for container in ax.containers:
    ax.bar_label(container, fontsize=12, fontweight='bold')

plt.title('Distribution of Survival', fontsize=14, fontweight='bold')
plt.xlabel('Survived (0 = No, 1 = Yes)', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.xticks([0, 1], ['Did Not Survive', 'Survived'])
plt.show()

# Insight
print("Value Counts:")
print(df['survived'].value_counts())
````

````python
# Count Plot for Sex
plt.figure(figsize=(8, 5))
ax = sns.countplot(data=df, x='sex', palette='Set2')

# Add count labels on bars
for container in ax.containers:
    ax.bar_label(container, fontsize=12, fontweight='bold')

plt.title('Distribution of Passenger Gender', fontsize=14, fontweight='bold')
plt.xlabel('Sex', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.show()

print("Value Counts:")
print(df['sex'].value_counts())
````

````python
# Count Plot for Passenger Class (Ordinal Category)
plt.figure(figsize=(10, 5))

# Define order for ordinal data (1st class is best, 3rd is lowest)
order = [1, 2, 3]
ax = sns.countplot(data=df, x='pclass', order=order, palette='viridis')

for container in ax.containers:
    ax.bar_label(container, fontsize=12, fontweight='bold')

plt.title('Distribution of Passenger Class', fontsize=14, fontweight='bold')
plt.xlabel('Passenger Class (1=1st, 2=2nd, 3=3rd)', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.show()

print("Value Counts:")
print(df['pclass'].value_counts().sort_index())
````

````python
# Count Plot for Embarked Port
plt.figure(figsize=(10, 5))

# Order by frequency for better visualization
order = df['embarked'].value_counts().index
ax = sns.countplot(data=df, x='embarked', order=order, palette='Set1')

for container in ax.containers:
    ax.bar_label(container, fontsize=12, fontweight='bold')

plt.title('Distribution of Embarkation Port', fontsize=14, fontweight='bold')
plt.xlabel('Port (C=Cherbourg, Q=Queenstown, S=Southampton)', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.show()

print("Value Counts:")
print(df['embarked'].value_counts())
````

## 2. Pie Chart

> A **pie chart** shows proportions as slices of a circle, making it easy to see percentages.

### Advantages:
- Visual representation of proportions
- Easy to understand percentages
- Great for communicating to non-technical audiences

### When to Use:
- **Best for:** 2-5 categories
- **Avoid when:** Many categories (use bar chart instead)

### Syntax:
```python
plt.pie(values, labels=labels, autopct='%1.1f%%')
```

````python
# Pie Chart for Survival and Sex
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Survival Pie Chart
survival_counts = df['survived'].value_counts()
colors_survival = ['#ff6b6b', '#4ecdc4']
labels_survival = ['Did Not Survive', 'Survived']
axes[0].pie(survival_counts, labels=labels_survival, autopct='%1.1f%%', 
            colors=colors_survival, explode=[0.02, 0.02], shadow=True,
            textprops={'fontsize': 12, 'fontweight': 'bold'})
axes[0].set_title('Survival Distribution', fontsize=14, fontweight='bold')

# Sex Pie Chart
sex_counts = df['sex'].value_counts()
colors_sex = ['#74b9ff', '#fd79a8']
axes[1].pie(sex_counts, labels=sex_counts.index, autopct='%1.1f%%',
            colors=colors_sex, explode=[0.02, 0.02], shadow=True,
            textprops={'fontsize': 12, 'fontweight': 'bold'})
axes[1].set_title('Gender Distribution', fontsize=14, fontweight='bold')

plt.tight_layout()
plt.show()
````

````python
# Pie Chart for Passenger Class
plt.figure(figsize=(8, 8))

class_counts = df['pclass'].value_counts().sort_index()
colors = ['#ffd93d', '#74b9ff', '#a29bfe']
labels = ['1st Class', '2nd Class', '3rd Class']

plt.pie(class_counts, labels=labels, autopct='%1.1f%%',
        colors=colors, explode=[0.02, 0.02, 0.02], shadow=True,
        textprops={'fontsize': 12, 'fontweight': 'bold'})
plt.title('Passenger Class Distribution', fontsize=14, fontweight='bold')
plt.show()

print("Passenger Class Breakdown:")
print(class_counts)
````

## 1. Histogram

> A **histogram** visualizes the distribution of numerical data by dividing values into "bins" (ranges) and counting how many values fall into each bin.

### How It Works:
```
Data: [18, 19, 19, 20, 20, 20, 21, 22, 23, 25]

Bins:     18-19  |  20-21  |  22-23  |  24-25
Count:      3    |    4    |    2    |    1

```

### Key Parameters:
- **`bins`**: Number of intervals (more bins = more detail)
- **`edgecolor`**: Border color of bars

### Syntax:

```python
import matplotlib.pyplot as plt

plt.hist(df['column'], bins=20, edgecolor='black')
plt.show()

````python
# Histogram for Age
plt.figure(figsize=(10, 6))

# Drop NaN values for Age
age_data = df['age'].dropna()

plt.hist(age_data, bins=20, edgecolor='black', color='#3498db', alpha=0.7)

# Add mean line
mean_age = age_data.mean()
plt.axvline(mean_age, color='red', linestyle='--', linewidth=2, label=f'Mean: {mean_age:.1f}')

# Add median line
median_age = age_data.median()
plt.axvline(median_age, color='green', linestyle='-.', linewidth=2, label=f'Median: {median_age:.1f}')

plt.title('Distribution of Passenger Age', fontsize=14, fontweight='bold')
plt.xlabel('Age (years)', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.legend()
plt.show()

print(f"Age Statistics:")
print(f"  Mean: {mean_age:.2f} years")
print(f"  Median: {median_age:.2f} years")
print(f"  Std: {age_data.std():.2f}")
print(f"  Range: {age_data.min():.0f} - {age_data.max():.0f} years")
print(f"  Missing Values: {df['age'].isna().sum()}")
````

````python
# Histogram for Fare - Exploring different bin sizes
fig, axes = plt.subplots(1, 3, figsize=(15, 4))

bin_sizes = [10, 30, 50]
colors = ['#e74c3c', '#9b59b6', '#1abc9c']

for ax, bins, color in zip(axes, bin_sizes, colors):
    ax.hist(df['fare'], bins=bins, edgecolor='black', 
            color=color, alpha=0.7)
    ax.set_title(f'bins = {bins}', fontsize=12, fontweight='bold')
    ax.set_xlabel('Fare (£)', fontsize=10)
    ax.set_ylabel('Frequency', fontsize=10)

plt.suptitle('Effect of Bin Size on Histogram Shape (Fare)', fontsize=14, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()

print("\n💡 Tip: Choosing the right number of bins is important!")
print("   Too few bins → Lose detail")
print("   Too many bins → Noisy, hard to see patterns")
````

## 2. Distribution Plot with KDE (Kernel Density Estimation)

> A **distplot** combines a histogram with a smooth **KDE curve** that shows the probability density.

### What is KDE?

**Kernel Density Estimation** creates a smooth curve that estimates the probability density function (PDF) of the data.

```
PDF (Probability Density Function):
- Y-axis represents DENSITY, not frequency
- Area under the curve = 1 (100%)
- Higher values = More likely to occur
```

### Advantages over Histogram:
- Smoother visualization
- Easier to compare distributions
- Reveals shape more clearly

### Syntax:
```python
sns.histplot(df['column'], kde=True)  # Modern seaborn
# or
sns.kdeplot(df['column'])  # KDE only
```

````python
# Distribution Plot for Age
plt.figure(figsize=(10, 6))

# Histogram with KDE (dropping NaN values)
age_data = df['age'].dropna()
sns.histplot(age_data, kde=True, color='#2ecc71', 
             edgecolor='black', alpha=0.7)

# Add mean line
mean_age = age_data.mean()
plt.axvline(mean_age, color='red', linestyle='--', linewidth=2, 
            label=f'Mean: {mean_age:.1f} years')

plt.title('Distribution of Passenger Age with KDE', fontsize=14, fontweight='bold')
plt.xlabel('Age (years)', fontsize=12)
plt.ylabel('Density / Count', fontsize=12)
plt.legend()
plt.show()

print("Interpretation:")
print(f"  Most passengers were around {mean_age:.0f} years old.")
print(f"  The KDE curve shows the probability density of passenger ages.")
print(f"  Notice the peak around 20-30 years - young adults were most common.")
````

````python
# KDE Plots for Multiple Numerical Variables
fig, axes = plt.subplots(2, 2, figsize=(12, 10))
axes = axes.flatten()

numerical_cols = ['age', 'fare', 'sibsp', 'parch']
colors = ['#3498db', '#e74c3c', '#2ecc71', '#9b59b6']
titles = ['Passenger Age', 'Ticket Fare (£)', 'Siblings/Spouses Aboard', 'Parents/Children Aboard']

for ax, col, color, title in zip(axes, numerical_cols, colors, titles):
    data = df[col].dropna()
    sns.histplot(data, kde=True, ax=ax, color=color, edgecolor='black', alpha=0.7)
    ax.set_title(f'Distribution of {title}', fontsize=11, fontweight='bold')
    ax.set_xlabel('')

plt.suptitle('Distribution of All Numerical Variables (Titanic)', fontsize=14, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()
````

## Understanding Skewness

> **Skewness** measures the asymmetry of the probability distribution.

---

### Types of Skewness

| Type | Skewness Value | Description | Example |
|------|----------------|-------------|---------|
| **Symmetric (Normal)** | ≈ 0 | Peak in center, equal tails | Heights, IQ scores |
| **Right-Skewed (Positive)** | > 0 | Tail extends right | Income, House prices |
| **Left-Skewed (Negative)** | < 0 | Tail extends left | Test scores (easy test) |

---

### Guide

```
Left-Skewed          Symmetric           Right-Skewed
(Negative)           (Normal)            (Positive)

Tail on LEFT        BALANCED            Tail on RIGHT

Mean < Median       Mean ≈ Median       Mean > Median
```

### Pandas Function:
```python
df['column'].skew()
# Returns: positive (right-skewed), negative (left-skewed), 0 (symmetric)
```

````python
# Calculate Skewness for all numerical columns
print("=" * 50)
print("SKEWNESS ANALYSIS")
print("=" * 50)

numerical_cols = ['age', 'fare', 'sibsp', 'parch']

for col in numerical_cols:
    skewness = df[col].skew()
    
    # Determine skewness type
    if abs(skewness) < 0.5:
        skew_type = "Approximately Symmetric"
    elif skewness > 0:
        skew_type = "Right-Skewed (Positive)"
    else:
        skew_type = "Left-Skewed (Negative)"
    
    print(f"\n{col.upper()}:")
    print(f"  Skewness: {skewness:.3f}")
    print(f"  Type: {skew_type}")

print("\n" + "=" * 50)
print("💡 Note: Fare is highly right-skewed - most paid low fares,")
print("   but a few wealthy passengers paid very high fares.")
````

## 3. Box Plot (Box and Whisker Plot)

> A **box plot** is a powerful visualization that displays the **five-number summary** and helps identify **outliers**.

---

### The Five-Number Summary

| Component | Definition | Percentile |
|-----------|------------|------------|
| **Minimum** | Q1 - 1.5 × IQR | Lower fence |
| **Q1** | First Quartile | 25th percentile |
| **Q2 (Median)** | Middle value | 50th percentile |
| **Q3** | Third Quartile | 75th percentile |
| **Maximum** | Q3 + 1.5 × IQR | Upper fence |

**IQR (Interquartile Range)** = Q3 - Q1

---

### Visual Anatomy of a Box Plot

```
                    Outliers
                       ●
                       │
        Maximum ───────┼─── (Q3 + 1.5 × IQR)
                       │
                   ┌───┴───┐
              Q3 ──┤       │── 75th Percentile
                   │       │
          Median ──┼───────┼── 50th Percentile (Q2)
                   │       │
              Q1 ──┤       │── 25th Percentile
                   └───┬───┘
                       │
        Minimum ───────┼─── (Q1 - 1.5 × IQR)
                       │
                       ●
                    Outliers
```

### Outlier Detection:
- Values **below** Q1 - 1.5 × IQR → Potential outliers
- Values **above** Q3 + 1.5 × IQR → Potential outliers

### Syntax:
```python
sns.boxplot(data=df, x='column')
# or for vertical
sns.boxplot(data=df, y='column')
```

````python
# Box Plot for Fare
plt.figure(figsize=(12, 5))

# Create subplot layout
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Horizontal Box Plot
sns.boxplot(data=df, x='fare', ax=axes[0], color='#3498db')
axes[0].set_title('Fare Distribution - Box Plot', fontsize=12, fontweight='bold')
axes[0].set_xlabel('Fare (£)', fontsize=11)

# Vertical Box Plot for comparison
sns.boxplot(data=df, y='fare', ax=axes[1], color='#e74c3c')
axes[1].set_title('Fare Distribution (Vertical)', fontsize=12, fontweight='bold')
axes[1].set_ylabel('Fare (£)', fontsize=11)

plt.tight_layout()
plt.show()

# Calculate statistics
Q1 = df['fare'].quantile(0.25)
Q2 = df['fare'].quantile(0.50)
Q3 = df['fare'].quantile(0.75)
IQR = Q3 - Q1

print("\nFive-Number Summary for Fare:")
print(f"  Minimum (Lower Fence): {max(df['fare'].min(), Q1 - 1.5 * IQR):.2f}")
print(f"  Q1 (25th Percentile):  £{Q1:.2f}")
print(f"  Q2 (Median):           £{Q2:.2f}")
print(f"  Q3 (75th Percentile):  £{Q3:.2f}")
print(f"  Maximum (Upper Fence): {min(df['fare'].max(), Q3 + 1.5 * IQR):.2f}")
print(f"  IQR:                   £{IQR:.2f}")
print(f"\n💡 Note: Many outliers exist (high fares for wealthy passengers)")
````

````python
# Box Plots for All Numerical Variables
fig, axes = plt.subplots(2, 2, figsize=(12, 10))
axes = axes.flatten()

numerical_cols = ['age', 'fare', 'sibsp', 'parch']
colors = ['#3498db', '#e74c3c', '#2ecc71', '#9b59b6']
titles = ['Age (years)', 'Fare (£)', 'Siblings/Spouses', 'Parents/Children']

for ax, col, color, title in zip(axes, numerical_cols, colors, titles):
    sns.boxplot(data=df, y=col, ax=ax, color=color)
    ax.set_title(f'{title}', fontsize=11, fontweight='bold')
    ax.set_ylabel('')

plt.suptitle('Box Plots: Identifying Outliers in Titanic Dataset', 
             fontsize=14, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()

print("\n💡 Interpretation:")
print("   Points outside the whiskers (●) are POTENTIAL OUTLIERS")
print("   - Fare: Many high-fare outliers (wealthy 1st class passengers)")
print("   - SibSp/Parch: Large families are rare outliers")
````

````python
# Detailed Outlier Detection
print("=" * 60)
print("OUTLIER DETECTION USING IQR METHOD")
print("=" * 60)

numerical_cols = ['age', 'fare', 'sibsp', 'parch']

for col in numerical_cols:
    data = df[col].dropna()
    Q1 = data.quantile(0.25)
    Q3 = data.quantile(0.75)
    IQR = Q3 - Q1
    
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    outliers = data[(data < lower_bound) | (data > upper_bound)]
    
    print(f"\n{col.upper()}:")
    print(f"  IQR Range: [{lower_bound:.2f}, {upper_bound:.2f}]")
    print(f"  Number of Outliers: {len(outliers)}")
    print(f"  Percentage: {len(outliers)/len(data)*100:.1f}%")
````

## General Best Practices

1. **Always identify data types first** - Categorical vs. Numerical determines your approach

2. **For Categorical Data:**
   - Use **count plots** for absolute frequencies
   - Use **pie charts** for proportions (best with 2-5 categories)

3. **For Numerical Data:**
   - Use **histograms** to see distribution shape
   - Use **KDE plots** for smooth probability density
   - Use **box plots** for outlier detection and 5-number summary

4. **Check for skewness** - It affects which statistical measures are appropriate:
   - Symmetric: Mean ≈ Median (use either)
   - Skewed: Median is more robust than Mean

5. **Identify outliers** - They can significantly impact your analysis and models

---

> 💡 **Pro Tip:** Univariate analysis is the foundation of EDA. Master this before moving to bivariate and multivariate analysis!
