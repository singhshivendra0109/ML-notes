# Exploratory Data Analysis: Bivariate & Multivariate Analysis

> **Definition:** Analyzing relationships between **two or more variables** to uncover patterns, trends, and insights in data.

---

## Types of Analysis

| Type | Variables | Description |
|------|-----------|-------------|
| **Bivariate** | 2 | Relationship between two columns |
| **Multivariate** | 3+ | Complex relationships across multiple columns |

---

## Data Type Combinations for Plotting

| Combination | Plot Types |
|-------------|------------|
| **Numerical-Numerical** | Scatter Plot, Line Plot, Pair Plot |
| **Numerical-Categorical** | Bar Plot, Box Plot, Dist Plot |
| **Categorical-Categorical** | Heat Map, Cluster Map |

````python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set visualization style
plt.style.use('seaborn-v0_8-whitegrid')
sns.set_palette('husl')

# Load datasets
tips = sns.load_dataset('tips')
titanic = sns.load_dataset('titanic')
iris = sns.load_dataset('iris')
flights = sns.load_dataset('flights')

print("Datasets loaded successfully!")
print(f"Tips: {tips.shape}, Titanic: {titanic.shape}, Iris: {iris.shape}, Flights: {flights.shape}")
````

# 1. Scatter Plot (Numerical to Numerical)

> Visualizes the **relationship between two numerical variables** as points on a 2D plane.

---

## When to Use
- Finding correlations between variables
- Identifying clusters or patterns
- Detecting outliers

## Syntax
```python
sns.scatterplot(data=df, x='column1', y='column2')
```

## Multivariate Parameters

| Parameter | Purpose | Data Type |
|-----------|---------|------------|
| `hue` | Color by category | Categorical |
| `style` | Marker style by category | Categorical |
| `size` | Point size by value | Numerical |

````python
# Basic Scatter Plot: Total Bill vs Tip
plt.figure(figsize=(10, 6))
sns.scatterplot(data=tips, x='total_bill', y='tip', alpha=0.7, s=80)

plt.title('Total Bill vs Tip Amount', fontsize=14, fontweight='bold')
plt.xlabel('Total Bill ($)', fontsize=12)
plt.ylabel('Tip ($)', fontsize=12)
plt.show()

# Calculate correlation
correlation = tips['total_bill'].corr(tips['tip'])
print(f"Correlation: {correlation:.3f}")
print("Insight: Higher bills generally lead to higher tips (positive correlation)")
````

````python
# Multivariate Scatter Plot (up to 5 dimensions of information)
plt.figure(figsize=(12, 8))

sns.scatterplot(
    data=tips,
    x='total_bill',
    y='tip',
    hue='sex',           # Color by gender
    style='smoker',      # Marker style by smoker status
    size='size',         # Point size by party size
    sizes=(50, 300),
    alpha=0.7
)

plt.title('Total Bill vs Tip (Multivariate Analysis)', fontsize=14, fontweight='bold')
plt.xlabel('Total Bill ($)', fontsize=12)
plt.ylabel('Tip ($)', fontsize=12)
plt.legend(bbox_to_anchor=(1.02, 1), loc='upper left')
plt.tight_layout()
plt.show()

print("This plot shows 5 pieces of information:")
print("  1. X-axis: Total Bill")
print("  2. Y-axis: Tip Amount")
print("  3. Color: Gender (Male/Female)")
print("  4. Style: Smoker Status (Yes/No)")
print("  5. Size: Party Size")
````

# 2. Bar Plot (Numerical to Categorical)

> Compares a **numerical quantity across different categories** using bars.

---

## When to Use
- Comparing means/averages across groups
- Showing aggregated values by category

## Syntax
```python
sns.barplot(data=df, x='categorical_col', y='numerical_col')
```
- **X-axis:** Categorical variable (discrete groups)
- **Y-axis:** Numerical variable (aggregated value like mean, sum, count)
- **Bars are separated:** Unlike histograms where bars touch, bar plot bars have gaps to show distinct categories

## Key Parameters

| Parameter | Purpose |
|-----------|---------|
| `estimator` | Function to aggregate (default: mean) |
| `hue` | Add another categorical dimension |
| `errorbar` | Show confidence intervals |

````python
# Bar Plot: Average Age by Passenger Class
plt.figure(figsize=(10, 6))
ax = sns.barplot(data=titanic, x='pclass', y='age', palette='viridis')

# Add value labels on bars
for container in ax.containers:
    ax.bar_label(container, fmt='%.1f', fontsize=11, fontweight='bold')

plt.title('Average Age by Passenger Class', fontsize=14, fontweight='bold')
plt.xlabel('Passenger Class', fontsize=12)
plt.ylabel('Average Age (years)', fontsize=12)
plt.xticks([0, 1, 2], ['1st Class', '2nd Class', '3rd Class'])
plt.show()

print("Insight: First class passengers were older on average")
print(titanic.groupby('pclass')['age'].mean().round(2))
````

````python
# Multivariate Bar Plot: Average Fare by Class and Gender
plt.figure(figsize=(10, 6))
ax = sns.barplot(data=titanic, x='pclass', y='fare', hue='sex', palette='Set2')

plt.title('Average Fare by Class and Gender', fontsize=14, fontweight='bold')
plt.xlabel('Passenger Class', fontsize=12)
plt.ylabel('Average Fare (£)', fontsize=12)
plt.xticks([0, 1, 2], ['1st Class', '2nd Class', '3rd Class'])
plt.legend(title='Gender')
plt.show()

print("Multivariate Analysis:")
print(titanic.groupby(['pclass', 'sex'])['fare'].mean().round(2))
````

# 3. Box Plot (Numerical to Categorical)

> Compares the **distribution of a numerical variable** across different categories.

---

## What It Shows
- Median (center line)
- Quartiles (box edges: Q1 and Q3)
- Whiskers (1.5 × IQR)
- Outliers (points beyond whiskers)

## Syntax
```python
sns.boxplot(data=df, x='categorical_col', y='numerical_col')
```

````python
# Box Plot: Age Distribution by Gender
plt.figure(figsize=(10, 6))
sns.boxplot(data=titanic, x='sex', y='age', palette='coolwarm')

plt.title('Age Distribution by Gender', fontsize=14, fontweight='bold')
plt.xlabel('Gender', fontsize=12)
plt.ylabel('Age (years)', fontsize=12)
plt.show()

print("Summary Statistics:")
print(titanic.groupby('sex')['age'].describe().round(2))
````

````python
# Multivariate Box Plot: Age by Gender and Survival
plt.figure(figsize=(10, 6))
sns.boxplot(data=titanic, x='sex', y='age', hue='survived', palette='Set1')

plt.title('Age Distribution by Gender and Survival', fontsize=14, fontweight='bold')
plt.xlabel('Gender', fontsize=12)
plt.ylabel('Age (years)', fontsize=12)
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.show()

print("Insight: Compare age distributions between survivors and non-survivors within each gender")
````

# 4. Distribution Plot (Numerical by Category)

> Plots the **Probability Density Function (PDF)** of a numerical variable, separated by categories.

---

## When to Use
- Compare distributions across groups
- Identify survival patterns by demographics

## Syntax Options

| Type | Syntax |
|------|--------|
| **Only KDE (smooth curve)** | `sns.kdeplot(data=df, x='col', hue='cat_col')` |
| **Only Histogram** | `sns.histplot(data=df, x='col', hue='cat_col')` |
| **Both KDE + Histogram** | `sns.histplot(data=df, x='col', hue='cat_col', kde=True)` |

- **X-axis:** Numerical variable (continuous values)
- **Y-axis:** Density (probability, not count)
- **Curves represent:** Smoothed distribution (KDE - Kernel Density Estimate)
- **Area under curve:** Always equals 1 (probability)
- **Use `hue` parameter:** To compare distributions across categorical groups

````python
# Distribution Plot: Age Distribution by Survival
plt.figure(figsize=(10, 6))

# Create separate KDE plots for survived vs not survived
sns.kdeplot(data=titanic[titanic['survived']==0], x='age', 
            label='Did Not Survive', color='#e74c3c', fill=True, alpha=0.3)
sns.kdeplot(data=titanic[titanic['survived']==1], x='age', 
            label='Survived', color='#2ecc71', fill=True, alpha=0.3)

# Using hue parameter (simpler approach)
# sns.kdeplot(data=titanic, x='age', hue='survived', fill=True, alpha=0.3, 
#             palette={0: '#e74c3c', 1: '#2ecc71'})

plt.title('Age Distribution: Survived vs Did Not Survive', fontsize=14, fontweight='bold')
plt.xlabel('Age (years)', fontsize=12)
plt.ylabel('Density', fontsize=12)
plt.legend()
plt.show()

print("Insight: Children (lower ages) had higher survival probability")
print("         Notice the peak for survivors at young ages")
````

# 5. Heat Map (Categorical to Categorical)

> Visualizes the **relationship between two categorical variables** using color intensity in a matrix.

---

## When to Use
- Show counts/percentages in cross-tabulation
- Correlation matrices
- Pivot table visualization

## Syntax Options

| Type | Syntax |
|------|--------|
| **Quick one-liner** | `sns.heatmap(pd.crosstab(df['col1'], df['col2']))` |
| **With customization** | Create crosstab first, then `sns.heatmap(crosstab, annot=True, cmap='YlOrRd')` |

## Syntax (With customization)
```python
# Create crosstab first
cross_tab = pd.crosstab(df['col1'], df['col2'])
sns.heatmap(cross_tab, annot=True, cmap='YlOrRd')
```
- **Crosstab:** Creates a frequency table between two categorical variables
- **Color intensity:** In a heat map, It represents the count/value in each cell. You can customize colors using the cmap (colormap) parameter.

## Key Parameters

| Parameter | Purpose |
|-----------|---------|
| `annot` | Display values in cells |
| `cmap` | Color palette |
| `fmt` | Number format |

````python
# Heat Map: Survival by Passenger Class
plt.figure(figsize=(8, 6))

# Create crosstab
survival_class = pd.crosstab(titanic['pclass'], titanic['survived'])
survival_class.columns = ['Did Not Survive', 'Survived']
survival_class.index = ['1st Class', '2nd Class', '3rd Class']

sns.heatmap(survival_class, annot=True, fmt='d', cmap='YlOrRd', 
            linewidths=0.5, cbar_kws={'label': 'Count'})

plt.title('Survival Count by Passenger Class', fontsize=14, fontweight='bold')
plt.xlabel('Survival Status', fontsize=12)
plt.ylabel('Passenger Class', fontsize=12)
plt.show()

print(survival_class)
````

````python
# Heat Map with Survival Percentages
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# By Class
survival_pct_class = pd.crosstab(titanic['pclass'], titanic['survived'], normalize='index') * 100
survival_pct_class.columns = ['Did Not Survive %', 'Survived %']
sns.heatmap(survival_pct_class, annot=True, fmt='.1f', cmap='RdYlGn', ax=axes[0])
axes[0].set_title('Survival Rate by Class (%)', fontweight='bold')

# By Gender
survival_pct_sex = pd.crosstab(titanic['sex'], titanic['survived'], normalize='index') * 100
survival_pct_sex.columns = ['Did Not Survive %', 'Survived %']
sns.heatmap(survival_pct_sex, annot=True, fmt='.1f', cmap='RdYlGn', ax=axes[1])
axes[1].set_title('Survival Rate by Gender (%)', fontweight='bold')

plt.tight_layout()
plt.show()

print("Insight: 1st class and females had significantly higher survival rates")
````

# 6. Cluster Map (Categorical to Categorical)

> Like a heat map but also **groups similar categories** together using dendrograms.

---

## What It Shows
- Heat map values
- Hierarchical clustering on rows and columns
- Dendrograms showing "closeness" of categories

## Syntax

### Two-step Approach
```python
# Step 1: Prepare data using pivot table
aggregated_data = df.pivot_table(
    values='<numeric_column>',      # Column to aggregate
    index='<row_variable>',         # Variable for rows
    columns='<column_variable>',    # Variable for columns
    aggfunc='<aggregation_func>'    # 'mean', 'sum', 'count', etc.
).fillna(0)

# Step 2: Create clustermap
sns.clustermap(
    aggregated_data,
    annot=True,                     # Show values in cells
    cmap='<colormap>',              # Color scheme (e.g., 'coolwarm', 'viridis')
    fmt='<format_string>',          # Number format (e.g., '.2f', 'd')
    figsize=(width, height)         # Figure dimensions
)
```
### One-liner
```python
sns.clustermap(
    pd.crosstab(df['<row_variable>'], df['<column_variable>']),
    annot=True,
    cmap='<colormap>'
)
```

````python
# Cluster Map: Family Size and Survival
# Create pivot table of survival rate by SibSp and Parch
family_survival = titanic.pivot_table(
    values='survived', 
    index='sibsp', 
    columns='parch', 
    aggfunc='mean'
).fillna(0)

# Create cluster map
g = sns.clustermap(family_survival, annot=True, fmt='.2f', cmap='RdYlGn', figsize=(10, 8),linewidths=0.5)

g.fig.suptitle('Survival Rate by Family Size (Clustered)', fontsize=14, fontweight='bold', y=1.02)
plt.show()

print("Dendrograms show which family size combinations have similar survival patterns")
print("Closer branches = more similar survival rates")
````

# 7. Pair Plot (Collection of Scatter Plots)

> Automatically generates **scatter plots for every pair** of numerical columns.
We can also color by a categorical column.
---

## What It Shows
- Grid of scatter plots (off-diagonal)
- Histograms/KDE on diagonal (variable vs itself)
- Quick overview of all pairwise relationships

## Syntax
```python
sns.pairplot(data=df, hue='categorical_col')
```

## Key Parameters

| Parameter | Purpose |
|-----------|---------|
| `hue` | Color by category |
| `diag_kind` | 'hist' or 'kde' for diagonal |
| `corner` | Show only lower triangle |

````python
# Pair Plot: Iris Dataset
g = sns.pairplot(iris, hue='species', palette='husl', 
                 diag_kind='kde', corner=False)

g.fig.suptitle('Iris Dataset: Pairwise Relationships', fontsize=14, fontweight='bold', y=1.02)
plt.show()

print("Insight: Pair plots help identify which variables best separate the species")
print("         Notice how petal_length and petal_width clearly separate species")
````

# 8. Line Plot (Numerical to Numerical - Time Based)

> A scatter plot with **points connected by lines**, ideal for **time-series data**.

---

## When to Use
- One variable is time-based (year, month, date)
- Showing trends over time
- Continuous progression

## Syntax
```python
sns.lineplot(data=df, x='time_col', y='value_col')
```

````python
# Line Plot: Passengers Over Years
plt.figure(figsize=(12, 6))

# Aggregate by year
yearly_passengers = flights.groupby('year')['passengers'].sum().reset_index()

sns.lineplot(data=yearly_passengers, x='year', y='passengers', 
             marker='o', markersize=8, linewidth=2, color='#3498db')

plt.title('Total Flight Passengers by Year', fontsize=14, fontweight='bold')
plt.xlabel('Year', fontsize=12)
plt.ylabel('Total Passengers', fontsize=12)
plt.grid(True, alpha=0.3)
plt.show()

print("Insight: Clear upward trend in flight passengers over the years")
````

````python
# Heat Map: Passengers by Month and Year (Pivot Table Visualization)
plt.figure(figsize=(12, 8))

# Create pivot table
flights_pivot = flights.pivot(index='month', columns='year', values='passengers')

sns.heatmap(flights_pivot, annot=True, fmt='d', cmap='YlOrRd', 
            linewidths=0.5, cbar_kws={'label': 'Passengers'})

plt.title('Flight Passengers by Month and Year', fontsize=14, fontweight='bold')
plt.xlabel('Year', fontsize=12)
plt.ylabel('Month', fontsize=12)
plt.show()

print("Insight: Summer months (Jun-Aug) show higher passenger counts")
print("         Seasonal travel patterns are clearly visible")
````

# Summary: Bivariate & Multivariate Analysis

---

## Quick Reference Table

| Combination | Plot Type | Use Case | Multivariate Extension |
|-------------|-----------|----------|------------------------|
| **Num-Num** | Scatter Plot | Correlation | hue, style, size |
| **Num-Num (Time)** | Line Plot | Trends over time | Multiple lines |
| **Num-Cat** | Bar Plot | Compare averages | hue parameter |
| **Num-Cat** | Box Plot | Compare distributions | hue parameter |
| **Num-Cat** | Dist Plot | Compare PDFs | Multiple KDEs |
| **Cat-Cat** | Heat Map | Cross-tabulation | Color intensity |
| **Cat-Cat** | Cluster Map | Grouped similarities | Dendrograms |
| **Multiple Num** | Pair Plot | All pairwise relations | hue for grouping |

---

## Key Takeaways

1. **Choose plot based on data types** - The combination determines the visualization

2. **Multivariate extensions:**
   - Use `hue` for categorical grouping (color)
   - Use `style` for marker styles
   - Use `size` for numerical scaling

3. **Heat maps & Cluster maps:**
   - Create crosstab/pivot tables first
   - Use `annot=True` to show values
   - Choose appropriate color maps

4. **Time-series:**
   - Line plots for trends
   - Heat maps for seasonal patterns

---

> 💡 **Pro Tip:** Always start with simple bivariate analysis before adding more dimensions. Too much information can obscure patterns!
