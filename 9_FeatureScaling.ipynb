# Feature Scaling

> **Definition:** A technique to bring **independent features** in a dataset to a **smaller, similar range**. For example, if "IQ" is in the range of 100 and "CGPA" is in the range of 10, feature scaling brings them to a similar range.

---

## Why is Feature Scaling Important?

Feature scaling is typically the **last step in feature engineering** before feeding data to a machine learning model.

| Problem | Example | Impact |
|---------|---------|--------|
| **Scale Dominance** | "Age" in tens, "Salary" in thousands | Larger scale features dominate distance calculations |
| **Gradient Convergence** | Different feature ranges | Slower or unstable convergence in gradient-based algorithms |
| **Distance Calculations** | KNN, K-Means | Features with larger scales get more weight |

---

## Types of Feature Scaling

| Type | Technique | Description |
|------|-----------|-------------|
| **Standardization** | Z-score Normalization | Transforms to mean=0, std=1 |
| **Normalization** | Min-Max Scaling | Scales to [0, 1] range |
| **Normalization** | Robust Scaling | Uses median and IQR (robust to outliers) |

# 2. Standardization (Z-score Normalization)

> Transforms features so they have a **mean of 0** and a **standard deviation of 1**.

---

Standardization does NOT change the shape of the data distribution.

It only changes:

Center → Mean becomes 0 (mean centering)

Spread → Standard deviation becomes 1 (scaling)

What Remains the Same:

If data was normally distributed → It stays normally distributed

If data was skewed → It stays skewed

If data had outliers → Outliers remain outliers

Relative positions of data points are preserved

---

## Formula

$$X_{scaled} = \frac{X - \mu}{\sigma}$$

Where:
- $X$ = Original feature value
- $\mu$ = Mean of the feature
- $\sigma$ = Standard deviation of the feature

---

## Geometric Interpretation

| Step | Operation | Effect |
|------|-----------|--------|
| **1. Mean Centering** | $X - \mu$ | Shifts data so mean is at origin |
| **2. Scaling** | Divide by $\sigma$ | Makes spread uniform across axes |

---

# 3. Impact of Outliers on Standardization

> **Important:** Standardization **does NOT reduce** the impact of outliers!

---

## Why Outliers Are Still Problematic

| Issue | Explanation |
|-------|-------------|
| **Mean is affected** | Outliers shift the mean towards them |
| **Std Dev is inflated** | Outliers increase the standard deviation |
| **Relative distance preserved** | Outliers maintain their extreme positions |

---

## Key Insight

After standardization, outliers **will still behave as outliers** in the scaled data. They maintain their relative distance from the main cluster of data points.

> 💡 **Pro Tip:** Handle outliers **before** applying standardization for better results!

````python
# Create dataset with outliers
np.random.seed(42)

# Normal data
normal_data = np.random.normal(50, 10, 100).reshape(-1, 1)

# Add outliers
outliers = np.array([[150], [160], [180]])  # Extreme outliers
data_with_outliers = np.vstack([normal_data, outliers])

# Standardize
scaler = StandardScaler()
scaled_data = scaler.fit_transform(data_with_outliers)

# Visualize
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Before
axes[0].boxplot(data_with_outliers, vert=True)
axes[0].scatter([1]*len(data_with_outliers), data_with_outliers, alpha=0.5, color='#3498db')
axes[0].set_title('Before Standardization', fontsize=14, fontweight='bold')
axes[0].set_ylabel('Value')

# After
axes[1].boxplot(scaled_data, vert=True)
axes[1].scatter([1]*len(scaled_data), scaled_data, alpha=0.5, color='#2ecc71')
axes[1].set_title('After Standardization', fontsize=14, fontweight='bold')
axes[1].set_ylabel('Scaled Value')

plt.suptitle('Outliers Remain Outliers After Standardization', fontsize=14, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()

print("Original Data Statistics:")
print(f"  Mean: {data_with_outliers.mean():.2f}")
print(f"  Std:  {data_with_outliers.std():.2f}")
print(f"  Max (outlier): {data_with_outliers.max():.2f}")
print("\nScaled Data Statistics:")
print(f"  Mean: {scaled_data.mean():.10f}")
print(f"  Std:  {scaled_data.std():.2f}")
print(f"  Max (outlier): {scaled_data.max():.2f} standard deviations away!")
print("\nConclusion: Outliers are STILL far from the main cluster!")
````

# 4. When to Use Standardization

> Standardization is crucial for **distance-based** and **gradient-based** algorithms.

---

## Algorithms That NEED Standardization

| Algorithm | Reason |
|-----------|--------|
| **K-Nearest Neighbors (KNN)** | Uses distance calculations between data points |
| **K-Means Clustering** | Uses distance to cluster centroids |
| **Principal Component Analysis (PCA)** | Aims to maximize variance; sensitive to scale |
| **Logistic Regression** | Uses gradient descent for optimization |
| **Neural Networks** | Gradient descent with multiple layers |
| **Support Vector Machines (SVM)** | Distance-based decision boundaries |

---

## Algorithms That DON'T Need Standardization

| Algorithm | Reason |
|-----------|--------|
| **Decision Trees** | Splits based on thresholds, not distances |
| **Random Forests** | Ensemble of decision trees |
| **Gradient Boosting (XGBoost, LightGBM)** | Tree-based splitting |
| **Naive Bayes** | Based on probability, not distance |

````python
# Demonstrate impact on KNN (Distance-based algorithm)
from sklearn.neighbors import KNeighborsClassifier

# Prepare data
X = df[['Age', 'EstimatedSalary']]
y = df['Purchased']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

# Scale the data - fit on train only, transform both
scaler = StandardScaler()
scaler.fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Compare KNN performance for different k values
k_values = [1, 3, 5, 7, 9, 11]
results = []

for k in k_values:
    # Without scaling
    knn_no_scale = KNeighborsClassifier(n_neighbors=k)
    knn_no_scale.fit(X_train, y_train)
    acc_no_scale = accuracy_score(y_test, knn_no_scale.predict(X_test))
    
    # With scaling
    knn_scaled = KNeighborsClassifier(n_neighbors=k)
    knn_scaled.fit(X_train_scaled, y_train)
    acc_scaled = accuracy_score(y_test, knn_scaled.predict(X_test_scaled))
    
    results.append({'k': k, 'Without Scaling': acc_no_scale, 'With Scaling': acc_scaled})

results_df = pd.DataFrame(results)
print("KNN Accuracy Comparison:")
print(results_df.to_string(index=False))

# Visualize
plt.figure(figsize=(10, 6))
plt.plot(results_df['k'], results_df['Without Scaling'], 'o-', label='Without Scaling', color='#e74c3c', linewidth=2)
plt.plot(results_df['k'], results_df['With Scaling'], 's--', label='With Scaling', color='#2ecc71', linewidth=2)
plt.title('KNN Performance: With vs Without Scaling', fontsize=14, fontweight='bold')
plt.xlabel('Number of Neighbors (k)', fontsize=12)
plt.ylabel('Accuracy', fontsize=12)
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
````

# 5. Other Scaling Techniques (Bonus)

> While standardization is most common, other scaling methods exist for specific use cases.

---

## Min-Max Normalization

Scales features to a **fixed range [0, 1]**.

$$X_{scaled} = \frac{X - X_{min}}{X_{max} - X_{min}}$$

| Pros | Cons |
|------|------|
| Guaranteed bounds [0, 1] | Sensitive to outliers |
| Good for image data | Can compress most data into small range |

---

## Robust Scaling

Uses **median** and **IQR** instead of mean and std (robust to outliers).

$$X_{scaled} = \frac{X - median}{IQR}$$

| Pros | Cons |
|------|------|
| Robust to outliers | Does not bound values |
| Better for skewed data | Less common |

````python
# Compare all three scaling methods
from sklearn.preprocessing import StandardScaler, MinMaxScaler, RobustScaler

# Sample data with outliers
np.random.seed(42)
sample = np.concatenate([np.random.normal(50, 10, 95), [120, 140, 160, 180, 200]])
sample_reshaped = sample.reshape(-1, 1)

# Apply different scalers
standard = StandardScaler().fit_transform(sample_reshaped)
minmax = MinMaxScaler().fit_transform(sample_reshaped)
robust = RobustScaler().fit_transform(sample_reshaped)

# Create comparison DataFrame
comparison = pd.DataFrame({
    'Original': sample,
    'StandardScaler': standard.flatten(),
    'MinMaxScaler': minmax.flatten(),
    'RobustScaler': robust.flatten()
})

# Visualize
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

axes[0, 0].boxplot(sample)
axes[0, 0].set_title('Original Data', fontsize=12, fontweight='bold')

axes[0, 1].boxplot(standard)
axes[0, 1].set_title('StandardScaler (Z-score)', fontsize=12, fontweight='bold')

axes[1, 0].boxplot(minmax)
axes[1, 0].set_title('MinMaxScaler [0, 1]', fontsize=12, fontweight='bold')

axes[1, 1].boxplot(robust)
axes[1, 1].set_title('RobustScaler (Median-IQR)', fontsize=12, fontweight='bold')

plt.suptitle('Comparison of Different Scaling Techniques', fontsize=14, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()

print("Statistics Comparison:")
print(comparison.describe().round(3))
````

# Summary: Feature Scaling

---

## Quick Reference Table

| Technique | Formula | When to Use |
|-----------|---------|-------------|
| **Standardization** | $(X - \mu) / \sigma$ | Most ML algorithms, especially gradient-based |
| **Min-Max Scaling** | $(X - X_{min}) / (X_{max} - X_{min})$ | Neural networks, image data |
| **Robust Scaling** | $(X - median) / IQR$ | Data with significant outliers |

---

## Algorithms & Scaling Need

| Needs Scaling ✅ | Doesn't Need Scaling ❌ |
|------------------|------------------------|
| KNN, K-Means | Decision Trees |
| PCA | Random Forest |
| Logistic Regression | Gradient Boosting |
| Neural Networks | Naive Bayes |
| SVM | |

---

## Key Takeaways

1. **Purpose:** Bring features to similar scales so all contribute equally

2. **Standardization preserves shape:** Only changes center and spread, not the distribution

3. **Outliers remain outliers:** Standardization does NOT reduce outlier impact

4. **Always use for:** Distance-based and gradient-based algorithms

5. **Process order:**
   - Handle outliers first (if needed)
   - Then apply feature scaling
   - Use `fit()` on training data only
   - Use `transform()` on both train and test data (to avoid data leakage)

---

> 💡 **Pro Tip:** Always apply feature scaling as the **last preprocessing step** before training your model!
